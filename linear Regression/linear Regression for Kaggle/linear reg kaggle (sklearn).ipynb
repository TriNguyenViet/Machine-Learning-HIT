```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd
import math
```

```python
data_train_set=pd.read_csv('train.csv')
data_test_set=pd.read_csv('test.csv')
print(data_train_set.shape)
print(pd.DataFrame(data_train_set).head(5))
```

```python
train_set=data_train_set.dropna()
print(train_set.shape)
test_set=data_test_set.dropna()
```

```python
X=train_set[['x']].as_matrix()
Y=train_set[['y']].as_matrix()
plt.scatter(X,Y,c='black')
plt.xlabel('training_set_X')
plt.ylabel('training_set_Y')
plt.show()
```

```python
one = np.ones((X.shape[0],1))
X = np.concatenate((X,one),axis=1)
lm=linear_model.LinearRegression()
lm.fit(X,Y)
```

```python
print(lm.coef_)
```

```python
theta=lm.coef_
linex = [i for i in range(0,105)]
liney = [i*theta[0,0]+ theta[0,1] for i in range(0,105)]
```

```python
plt.scatter(X[:,0],Y,s=5,c='black')
plt.xlabel('training_set_X')
plt.ylabel('training_set_Y')
plt.scatter(linex,liney)
plt.show()
```

```python
X_test=test_set[['x']].as_matrix()
Y_test=test_set[['y']].as_matrix()
for a,b in zip(X_test[:],Y_test[:]):
    print('x test: {0}, thuc te : {1}, du doan : {2}'.format(a,b,a*theta[0,0]+theta[0,1]))
```

```python
plt.title('plot between actual values and predicted values in the test set')
plt.scatter(Y_test,X_test*theta[0,0]+theta[0,1],s=9,c='cyan')
plt.xlabel('test values')
plt.ylabel('predicted values')
plt.show()
```
